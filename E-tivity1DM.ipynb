Distribution Analysis

```python
# Let's start by plotting a histogram of ApplicantIncome using the command:

df['age'].hist(bins=50)
plt.xlabel('age 50 bins')
plt.ylabel('Number')
plt.show()

df['age'].hist(bins=25)
plt.xlabel('age 25 bins')
plt.ylabel('Number')
plt.show()

df['balance'].hist(bins=50)
plt.xlabel('balance')
plt.ylabel('Number')
plt.show()

df['day'].hist(bins=50)
plt.xlabel('day')
plt.ylabel('Number')
plt.show()

df['duration'].hist(bins=50)
plt.xlabel('duration')
plt.ylabel('Number')
plt.show()

df['campaign'].hist(bins=50)
plt.xlabel('campaign')
plt.ylabel('Number')
plt.show()

df['pdays'].hist(bins=50)
plt.xlabel('pdays')
plt.ylabel('Number')
plt.show()

df['previous'].hist(bins=50)
plt.xlabel('previous')
plt.ylabel('Number')
plt.show()
```

Plotting the histograms of the numerical component illustrates a number of interesting things:

1. The number of customers drops significantly between the ages of 60 and 70. Obviously old age and retirement and life expectancy play a part in this but nevertheless the slope of that reduction is interesting. The age is a skewed normal distribution.
2. The vast majority of people have less than 17,000 in their bank account. They always say you should 6 months salary in a bank account for a rainy day and these data suggests that this may not always be the case.This is a skewed distribution.
3. The Day numeric vlues distribution plot has a degree of bimodality associated with it and this might waarrant further investigation
4. Duration is a skewed normal distribution whilst number is simply a kewed distribution.


Given the number of normal distribution it is interesting that doesnt manifest itself in good correlations between vriables and suggests a high degree of independence between them.



Next we wil look a bit more box plots of balance grouped by attribute

```python
# Boxp lots of ApplicantIncome, grouped by the categorical attribute Education
df.boxplot(column='balance', by='marital')
plt.title('Box plot of Balance grouped by Marital Status')
plt.suptitle("") # get rid of the automatic 'Box plot grouped by group_by_column_name' title
plt.show()

df.boxplot(column='balance', by='loan')
plt.title('Box plot of Balance grouped by loan status')
plt.suptitle("") # get rid of the automatic 'Box plot grouped by group_by_column_name' title
plt.show()

df.boxplot(column='balance', by='education')
plt.title('Box plot of Balance grouped by Education')
plt.suptitle("") # get rid of the automatic 'Box plot grouped by group_by_column_name' title
plt.show()

df.boxplot(column='balance', by='housing')
plt.title('Box plot of Balance grouped by housing')
plt.suptitle("") # get rid of the automatic 'Box plot grouped by group_by_column_name' title
plt.show()

df.boxplot(column='balance', by='default')
plt.title('Box plot of Balance grouped by default')
plt.suptitle("") # get rid of the automatic 'Box plot grouped by group_by_column_name' title
plt.show()
```

This is an interesting part of the analysis.
The first question i would be interested in exploring is how your bank balance can be affected by your marital status:

1. This box plot is interseting. Notwithstanding outlier's and the fact that there are different smaller sample sizes in each of the categories, it is clear the range of balances is smaller for divorced customers is lower than that of both married and divorced customers. I'm not sure what the underlying philisophical point to all of this, but it is indication that getting divorced can seriously affect your bank balance. It would be interesting to compare to this to data from differnt juristictions with diffrent divorce legislation to see if this was a contributing factor.
2. From the box plot of age versus education, there are a number of interesting features. The fraction of customers who attended primary school is clearly quite small relative to those with secondary and tertiary education and their range of balances is clearly lower than the other two categories. It would be interesting to see the age profile of primary educated customers as this migh illustrate changing education patterns and how this influences.
